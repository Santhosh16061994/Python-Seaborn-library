```python
# house price data set

from sklearn.datasets import load_boston # it is in numpy array
house_boston = load_boston()  

house = pd.DataFrame(house_boston.data, columns=house_boston.feature_names)
# se we need to change to pd datframe

house['PRICE'] = house_boston.target




```

```python
house.head()
```

```python
house.columns
```

```python
# we are using ditribution and corealtion matric plot
```

**Distribution** Plot

```python
# gives the range of data
# distribution of price
# magnitude or range
```

```python
sns.displot(house['PRICE'])
```

```python
sns.distplot(house['PRICE'])
# we will get probability curve
# gives max data in that perticular range
```

Heat Map

```python
from scipy.spatial.distance import correlation
correlation = house.corr()
```

```python
# constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':10}, cmap='Blues')



# colour ligh  - ve corelate dark + ve corelate
# f
#
#
#
```
