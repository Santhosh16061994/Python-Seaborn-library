# Bar Plot


```python
fig = plt.figure() # it will create empty plot in that we have add several thing
# in variable fig we are storing the plot
ax = fig.add_axes([0,0,1,1]) # ax refers axis , it includees our plt in rectangle 
# 00 represent Qudrrant 11 and represent height width gnerall represents area
# creating emplty fig and adding it.
languages = ['english','french','spanish','latin','german']
people = [100,50,70,84,56]
ax.bar(languages,people)
plt.xlabel('Lang')
plt.ylabel('No.ppl')
plt.show()
# when we 100 or more data point its useful
# tells how mant catergory are there and how many ppl in that catg.
# we chk the magnitude

```

# Pie Cahrt

```python
fig1 = plt.figure()
ax = fig1.add_axes([3,3,3,3]) # ax refers axis , it includees our plt in rectangle 
languages = ['english','french','spanish','latin','german']
people = [100,50,70,84,56]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

# Scatterplot

```python
x = np.linspace(0,10,30)
y = np.sin(x) # it will take x as angel and find the sign value for all x value
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1]) 
ax.scatter(x,y , color='g')
ax.scatter(x,z , color='b')
plt.show()

```

3d scatter plot

```python
from matplotlib import projections
fig3 = plt.figure()
ax = plt.axes(projection='3d')  # ax = fig2.add_axes([0,0,1,1]) rectangle is 2d form
# projection cmap
z= 20*np.random.random(100)
x= np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')

plt.show()

```
